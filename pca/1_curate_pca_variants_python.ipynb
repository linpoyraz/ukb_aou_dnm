```python
# for aou variants
valid_vids = set(passes['vid'])
# determine which vid to use for each row. 
# we do this because ref and alt don't match across ukb and aou in some cases 

def get_valid_vid(row):
    if row['vid'] in valid_vids:
        return row['vid']
    elif row['vid2'] in valid_vids:
        return row['vid2']
    else:
        return None

for i in range(1,23):
    f_all = f'./ukb_dragen/chr{i}_ukb_biallelic_passing.tsv.gz'  
    colnames = ['chr', 'pos', 'allele_id', 'ref', 'alt', 'qual', 'filter']
    c = 1
    for difs_table in pd.read_csv(f_all, sep='\t', chunksize=10000000, names=colnames, header=0):
        print(c)
        difs_table['chr'] = difs_table['chr'].astype(str)
        difs_table['pos'] = difs_table['pos'].astype(int).astype(str)
        difs_table['ref'] = difs_table['ref'].astype(str)
        difs_table['alt'] = difs_table['alt'].astype(str)
        difs_table['vid'] = "chr" + difs_table['chr'] + "-" + difs_table['pos'] + "-" + difs_table['ref'] + "-" + difs_table['alt']
        difs_table['vid2'] = "chr" + difs_table['chr'] + "-" + difs_table['pos'] + "-" + difs_table['alt'] + "-" + difs_table['ref']
        # filter where either vid or vid2 is in the valid set
        difs_table_filtered = difs_table[
        difs_table['vid'].isin(valid_vids) | difs_table['vid2'].isin(valid_vids)
        ].copy()
        difs_table_filtered['selected_vid'] = difs_table_filtered['vid'].where(
        difs_table_filtered['vid'].isin(valid_vids),
        difs_table_filtered['vid2']
        )
        # write to file (just the selected_vid column, or include other columns as needed)
        difs_table_filtered.to_csv(f'./pca/chr{i}_pca_variants_ukb_biallelic.part{c}.tsv', sep='\t', index=False)
        c += 1
```

```python
# take the filter-passing variants in ukb and modify such that the reference allele matches hg38

hl.get_reference('GRCh38').add_sequence(
    fasta_file= f'{bucket}/pca/Homo_sapiens_assembly38.fasta',
    index_file= f'{bucket}/pca/Homo_sapiens_assembly38.fasta.fai'
)

variants = hl.import_table(f'{bucket}/pca/hq_snps_aou_hgdp_1000g_ukbb_DRAGEN_pruned_variants.tsv', no_header=True)  # or .csv, etc.
variants = variants.rename({'f0': 'raw_id'})  # rename the single column

# parse the dragen-style id
variants = variants.annotate(
    # split by ":" → ["dragen", "chr1", "930248", "g", "a"]
    fields = variants.raw_id.split(":"),
    chr = variants.raw_id.split(":")[1],
    pos = hl.int32(variants.raw_id.split(":")[2]),
    ref = variants.raw_id.split(":")[3],
    alt = variants.raw_id.split(":")[4]
)

# construct initial locus and alleles
variants = variants.annotate(
    locus = hl.locus(variants.chr, variants.pos, reference_genome='GRCh38'),
    alleles = [variants.ref, variants.alt],
    swapped_alleles = [variants.alt, variants.ref]
)

# define reference allele at locus
ref_genome = hl.get_reference('GRCh38')
ref_base = hl.get_sequence(
    variants.locus.contig,
    variants.locus.position,
    reference_genome='GRCh38',
    before=0,
    after=0
)

# add match status
variants = variants.annotate(
    ref_matches = ref_base.lower() == variants.ref.lower(),
    swapped_ref_matches = ref_base.lower() == variants.alt.lower()
)

# normalize by swapping if needed
variants = variants.annotate(
    final_alleles = hl.if_else(
        variants.ref_matches, variants.alleles,
        hl.if_else(variants.swapped_ref_matches, variants.swapped_alleles, hl.missing(hl.tarray(hl.tstr)))
    )
)

# keep only valid variants
variants = variants.filter(hl.is_defined(variants.final_alleles))

# final locus + alleles, key by them
variants = variants.annotate(
    final_locus = variants.locus
)
variants = variants.key_by(locus=variants.final_locus, alleles=variants.final_alleles)

variants.write(f'{bucket}/pca/pca_variants.ht', overwrite = True)
```

```python
library(data.table)
library(dplyr)
```

```python
variants = fread("./pca/pca_variants_acaf.tsv", header=TRUE)
loadings = fread("./pca/hgdp_1000g_variant_loadings_acaf_af_added.tsv", header=TRUE)
```

```python
swapped = 0
for (i in 1:nrow(loadings)){
    # if final_alleles doesn't match alleles in variants 
    if (variants$final_alleles[i] != variants$alleles[i]){
        # swap alleles
        loadings$alleles[i] = variants$alleles[i]
         # then, multiply loadings by -1 
        loadings$loadings[i] = toJSON(fromJSON(loadings$loadings[i]*(-1)))
        # and do 1-af 
        loadings$af[i] = 1-loadings$af[i]
        swapped = swapped + 1
    }
}
```

```python
fwrite(list(variants$raw_id), "./pca/pca_variants_acaf_DRAGEN.tsv", sep = "\t", 
      row.names=FALSE, quote = FALSE, col.names = FALSE)
```

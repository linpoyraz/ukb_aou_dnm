```r
library(data.table)
library(ggplot2)
library(dplyr)
library(tidyr)
library(stringr)
```

```r
# read in metadata 
meta = fread("./pca/gnomad_meta_updated.tsv")
# read in pc scores 
pcs = fread("./pca/hgdp_1000g_scores_acaf.tsv")
```

```r
pcs_clean = pcs %>%
  # remove the square brackets
  mutate(scores = str_remove_all(scores, "\\[|\\]")) %>%
  # separate into 20 new columns
  separate(scores, into = paste("pc", 1:20, sep = ""), sep = ",", convert = TRUE)
pcs_meta = merge(pcs_clean, meta, by = "s")
```

```r
#read in aou
aou = fread('./pca/aou_pcs_biallelic_0.tsv', sep = "\t")
for (i in 1:9){
    aou_part = fread(paste('./pca/aou_pcs_biallelic_', i, '.tsv', sep = ""), sep = "\t")
    aou = rbind(aou, aou_part)
}
```

```r
# read in ukbb
ukbb = fread('./pca/ukb_pcs_0_biallelic.tsv', sep = "\t")
for (i in 1:9){
    ukbb_part = fread(paste('./pca/ukb_pcs_', i, '_biallelic.tsv', sep = ""), sep = "\t")
    ukbb = rbind(ukbb, ukbb_part)
}
```

```r
aou_clean = aou %>%
  # remove the square brackets
  mutate(scores = str_remove_all(scores, "\\[|\\]")) %>%
  # separate into 20 new columns
  separate(scores, into = paste("pc", 1:20, sep = ""), sep = ",", convert = TRUE)
```

```r
ukbb_clean = ukbb %>%
  # remove the square brackets
  mutate(scores = str_remove_all(scores, "\\[|\\]")) %>%
  # separate into 20 new columns
  separate(scores, into = paste("pc", 1:20, sep = ""), sep = ",", convert = TRUE)
```

```r
fwrite(pcs_meta, "./pca/hgdp_1kg_pcs_biallelic.txt", sep = "\t", 
      row.names= FALSE, quote = FALSE)
fwrite(aou_clean, "./pca/aou_pcs_biallelic.txt", sep = "\t", 
      row.names= FALSE, quote = FALSE)
fwrite(ukbb_clean, "./pca/ukb_pcs_biallelic.txt", sep = "\t", 
      row.names= FALSE, quote = FALSE)
```

```python
import pandas as pd
```

```python
inds = {}

with open('/mnt/project/DNM/inds.txt') as f:
    
    for row in f:
        
        row = row.strip()
        row = row.split(' ')
        inds[int(row[1])] = True
```

```python
dnm_inds = {}
rel_inds = {}

with open('/mnt/project/DNM/twins_inds.txt') as f:
    
    for row in f:
        
        row = row.strip()
        row = row.split(' ')
        dnm_inds[int(row[1])] = True
        rel_inds[int(row[1])] = True
        
with open('/mnt/project/DNM/trios_inds.txt') as f:
    
    for row in f:
        
        row = row.strip()
        row = row.split(' ')
        dnm_inds[int(row[1])] = True
        rel_inds[int(row[1])] = True
        
with open('/mnt/project/DNM/sibs_inds.txt') as f:
    
    for row in f:
        
        row = row.strip()
        row = row.split(' ')
        dnm_inds[int(row[1])] = True
        rel_inds[int(row[1])] = True
```

```python
# Kinship data
inds_kin_df = pd.read_csv('/mnt/project/DNM/snipar/king.kin0', sep = '\t')
inds_kin_df.head()
```

```python
for _, row in inds_kin_df.iterrows():

    if int(row['ID1']) in dnm_inds or int(row['ID2']) in dnm_inds:
        rel_inds[int(row['ID1'])] = True
        rel_inds[int(row['ID2'])] = True
```

```python
unrel_inds = {ind: True for ind in inds if ind not in rel_inds}
```

```python
with open('unrelated_inds.txt', 'w') as f:
    for ind in unrel_inds:
        f.write(f'{ind} {ind}\n')
```

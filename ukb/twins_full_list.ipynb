```python
chrom = 1
blocks = 2
```

```python
inds_diffs = {}

path = f'/mnt/project/DNM/twins/diffs (full)/chr{chrom}/'

for b in range(blocks):

    with open(f'/mnt/project/DNM/twins/blocks/twins_pairs_b{b}.txt') as F:
        
        for i, _ in enumerate(F):
            
            # Remove pair if twin is missing
            cnt = 0
            
            with open(f'{path}twins_chr{chrom}_b{b}_p{i}.tsv') as f:
                
                for row in f:
                    
                    row = row.strip()
                    row = row.split()
                    cnt += 1 if len(row) > 0 else 0
                    
            if cnt < 2:
                continue
                
            # Record differences per twin
            with open(f'{path}twins_chr{chrom}_b{b}_p{i}.tsv') as f:
                
                for row in f:
                    
                    row = row.strip()
                    row = row.split()

                    ind_diffs = row[1].split('|')[1:] if len(row) == 2 else []
                    
                    for ind_diff in ind_diffs:
                        inds_diffs[ind_diff] = True
```

```python
macs = {}

path = f'/mnt/project/DNM/MAC (full)/'

with open(f'{path}macs_chr{chrom}.acount') as f:
    
    next(f)
    
    for row in f:
        
        row = row.strip()
        row = row.split()
        
        diff_pos = row[1].split(':')[2]
        diff_id = f'{row[0]}:{diff_pos}:{row[2]}:{row[3]}'
        diff_mac = int(row[4])
        diff_obs = int(row[5])
        
        if diff_id in inds_diffs:
            macs[diff_id] = (diff_mac, diff_obs)
```

```python
with open(f'twins_unrelated_chr{chrom}.txt', 'a') as f:
    for diff_id in inds_diffs:
        if diff_id in macs:
            f.write(f'{diff_id} {macs[diff_id][0]} {macs[diff_id][1]}\n')
```
